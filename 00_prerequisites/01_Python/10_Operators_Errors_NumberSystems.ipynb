# Operators in Python

```python
# Arithmetic Operators
a = 10
b = 3
```

```python
print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.33
print(a % b)   # 1
print(a ** b)  # 1000
print(a // b)  # 3
```

```python
# Comparison Operators
print(a > b)   # True
print(a == b)  # False
```

```python
# Logical Operators
x = True
y = False
```

```python
print(x and y)  # False
print(x or y)   # True
print(not x)    # False
```

```python
# Assignment Operators
a += 2  # Now a is 12
print(a)
```

```python
# Identity Operators
print(a is b)       # False
print(a is not b)   # True
```

```python
# Membership Operators
mylist = [1, 2, 3]
print(2 in mylist)      # True
print(4 not in mylist)  # True
```

# Number Systems & Conversions

```python
# Decimal to Other Number Systems
num = 25
print(bin(num))  # 0b11001 ‚Üí binary
print(oct(num))  # 0o31 ‚Üí octal
print(hex(num))  # 0x19 ‚Üí hexadecimal
```
